```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df_all = pd.read_csv('data/factor_grps.csv')
df = df_all.query('factor_hash=="factor0"')
df
```

```python
pd.ols(y=df.sharpe_ratio, x=df.stock_group)
```

```python
sns.lmplot("stock_group", "sharpe_ratio", df) 
```

```python
def get_factor_regression_line(df, factor_id):
    factor = 'factor'+str(factor_id)
    df = df_all.query('factor_hash=="%s"'%factor)
    sns.lmplot("stock_group", "sharpe_ratio", df) 
```

```python
for i in range(10):
    get_factor_regression_line(df_all, i)
```
