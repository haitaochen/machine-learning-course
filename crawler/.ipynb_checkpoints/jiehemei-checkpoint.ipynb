```python
import requests
from bs4 import BeautifulSoup
head = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:41.0) Gecko/20100101 Firefox/41.0'}
res = requests.get("http://www.cxjhm.com/forum-43-1.html", headers = head)
# print res.text
soup = BeautifulSoup(res.text)
table1 = soup.select("#threadlisttableid")
# print table1[0]
if len(table1)>0:
    for tbody1 in table1[0].select('tbody'):
        if len(tbody1.select(".s.xst"))>0:
            print tbody1.select(".s.xst")[0].text.strip()
            print "http://www.cxjhm.com/" + tbody1.select(".s.xst")[0].attrs["href"].strip()
        if len(tbody1.select(".by")[0].select('cite'))>0:
            if len(tbody1.select(".by")[0].select('cite'))>0:
                print tbody1.select(".by")[0].select('cite')[0].text.strip()
        if len(tbody1.select(".by")[0].select('em'))>0:
            if len(tbody1.select(".by")[0].select('em')[0].select('span'))>0:
                if len(tbody1.select(".by")[0].select('em')[0].select('span')[0].select('span'))>0:
                    print tbody1.select(".by")[0].select('em')[0].select('span')[0].select('span')[0].attrs['title'].strip()
            print tbody1.select(".by")[0].select('em')[0].text.strip()
        if len(tbody1.select(".num"))>0:
            if len(tbody1.select(".num")[0].select('a'))>0:
                print tbody1.select(".num")[0].select('a')[0].text.strip()
            if len(tbody1.select(".num")[0].select('em'))>0:    
                print tbody1.select(".num")[0].select('em')[0].text.strip()    
        print "--------------------------------------------"
```

```python
import sqlite3
cx = sqlite3.connect("jiehemei.db")
cu = cx.cursor()
cu.execute("create table tiezi(id integer primary key AutoIncrement,title varchar(100),content varchar(10000),source_url varchar(100), author varchar(100),publish_date date,huifu int,chakan int)")
cu.commit()
cu.close()
```
