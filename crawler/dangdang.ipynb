```python
import requests
from bs4 import BeautifulSoup
res = requests.get("http://search.dangdang.com/?key=java&SearchFromTop=1&catalog=")
# print res.text
soup = BeautifulSoup(res.text)
mainarea = soup.select(".con.shoplist")
for book in mainarea[0].select('li'):
#     print book.findAll(attrs={"name":"itemlist-title"})
#     print book.select(".pic")[0]
    print "书名:",book.select(".pic")[0].attrs['title'].strip()
    print "现价:",book.select(".search_now_price")[0].text.strip()
    print "原价:",book.select(".search_pre_price")[0].text.strip()
    authors = book.select(".search_book_author")
    print "作者:",authors[0].select('span')[0].text.strip()
    print "出版日期:",authors[0].select('span')[1].text.strip().replace("/","")
    print "出版社:",authors[0].select('span')[2].text.strip().replace("/","")
    print "评论数量:",book.select(".search_comment_num")[0].text.strip()
#     author1s = authors[0].findAll(attrs={"name":"itemlist-author"})
#     for author in author1s:
#         print author.text
    print "---------------------------------------------------------"
```
